```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load the model
model = YOLO("runs/pose/train/weights/best.pt")  # Load nano

# Predict with the model
results = model(
    "../datasets/Syringe-volume-estimation/test/images/IMG_4359_jpeg.rf.ab9b9b51520ac3b963519e4b1a8eb060.jpg",
    conf=0.5
    )

# Extract the image with annotations
annotated_image = results[0].plot()  # Visualize the results on the image

# Convert the image to RGB format (OpenCV uses BGR by default)
annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(annotated_image)
plt.axis("off")  # Turn off axes for better visualization
plt.show()
```
